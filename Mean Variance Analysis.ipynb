# Mean(Expected-Return)-Volatlity Analysis 

```python
import pandas as pd
import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt
```

```python
spy = yf.Ticker('pej')
```

```python
df = spy.history('120mo')
```

```python
df
```

```python
df_annual = df.resample('AS').first()
df_annual
```

```python
df_annual['YoY Return'] = df_annual['Close'].pct_change(periods=1) * 100
```

```python
df2 = df_annual
```

```python
df3 = df2.dropna()
```

```python
df3['YoY Return'].std()
```

```python
means = np.array([10.307045586546518,9.965300873421102,14.829421705189777,6.25620689959004,5.944517141112331,4.360671508273471])
vol = np.array([6.24213189646226,13.654618922198646,22.51742034295097,19.606518130252724,26.547029564738825,16.451586560903422])
Tickers = np.array(['Spy','Dia','Qqq','Vgk','Mchi','Pej'])
```

```python
type(means)
```

```python
df5 = pd.DataFrame({'Expected Return': means, 'Volatility': vol, 'Tickers': Tickers})
```

```python
df5['Sharpe-Ratio'] = df5['Expected Return'] / df5['Volatility']
```

```python
df5['Sharpe-Ratio'].sum()
```
